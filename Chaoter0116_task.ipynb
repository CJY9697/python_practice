__[문제 2: 붕어빵 주문 받기]__

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
p_count, s_count, c_count = "팥붕어빵 갯수", "슈크림붕어빵 갯수", "초코붕어빵 갯수"
order = {"팥붕어빵": 0, "슈크림붕어빵": 0, "초코붕어빵": 0}

주문 = input("주문하실 상품의 번호를 선택해주세요. (1. 팥붕어빵, 2.슈크림붕어빵, 3. 초코붕어빵")
if 주문 == "1":
    p_count = int(input("원하시는 개수를 입력해주세요."))
    order["팥붕어빵"] = p_count
    print(f"팥붕어빵을 {p_count}개 선택하셨습니다.")
elif 주문 == "2":
    s_count= int(input("원하시는 개수를 입력해주세요."))
    order["슈크림붕어빵"] = s_count
    print(f"슈크림붕어빵을 {s_count}개 선택하셨습니다.")
elif 주문 == "3":
    c_count = int(input("원하시는 개수를 입력해주세요."))
    order["초코붕어빵"] = c_count
    print(f"초코붕어빵을 {c_count}개 선택하셨습니다.")
else:
    print("숫자 1,2,3 중에 하나를 눌러주세요.")
```

__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
재고 = {"팥붕어빵": 10, "슈크림붕어빵": 8, "초코붕어빵": 5}
order = {"팥붕어빵": 0, "슈크림붕어빵": 0, "초코붕어빵": 0}
menu = {"1": "팥붕어빵", "2": "슈크림붕어빵", "3": "초코붕어빵"}
주문 = input("주문하실 상품의 번호를 선택해주세요. (1. 팥붕어빵, 2. 슈크림붕어빵, 3. 초코붕어빵)")
if 주문 in menu:
    붕어빵 = menu[주문]
    개수 = int(input(f"{붕어빵}의 원하시는 개수를 입력해주세요. "))
    
    if 개수 > 재고[붕어빵]:
        print(f"죄송합니다. {붕어빵}의 재고가 부족합니다.")
    else:
        재고[붕어빵] -= 개수
        order[붕어빵] = 개수
        print(f"{붕어빵} {개수}개를 구매하셨습니다.")
```

__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
재고 = {"팥붕어빵": 10, "슈크림붕어빵": 8, "초코붕어빵": 5}
order = {"팥붕어빵": 0, "슈크림붕어빵": 0, "초코붕어빵": 0}
menu = {"1": "팥붕어빵", "2": "슈크림붕어빵", "3": "초코붕어빵"}
while True:
    주문 = input("주문하실 상품의 번호를 선택해주세요. (1. 팥붕어빵, 2. 슈크림붕어빵, 3. 초코붕어빵, 4.종료)")
    if 주문 == "4":
        print("감사합니다.")
        break
    if 주문 in menu:
        붕어빵 = menu[주문]
        개수 = int(input(f"{붕어빵}의 원하시는 개수를 입력해주세요. "))
    if 개수 > 재고[붕어빵]:
        print(f"죄송합니다. {붕어빵}의 재고가 부족합니다.")
    else:
        재고[붕어빵] -= 개수
        order[붕어빵] += 개수
        print(f"{붕어빵} {개수}개를 구매하셨습니다.")
        
print(f"현재 붕어빵 재고 상황은 {재고}입니다.")
```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
재고 = {"팥붕어빵": 10, "슈크림붕어빵": 8, "초코붕어빵": 5}
new = {"팥붕어빵": 0, "슈크림붕어빵": 0, "초코붕어빵": 0}
menu = {"1": "팥붕어빵", "2": "슈크림붕어빵", "3": "초코붕어빵"}

while True:
    주문 = input("주문하실 상품의 번호를 선택해주세요. (1. 팥붕어빵, 2. 슈크림붕어빵, 3. 초코붕어빵, 4. 종료)")
    
    if 주문 == "4":
        print(f"총 추가된 붕어빵 개수: {new}")
        print(f"현재 붕어빵 재고 상황은 {재고}입니다.")
        print("재고 추가를 종료합니다.")
        break

    if 주문 in menu:
        붕어빵 = menu[주문]
        개수 = int(input(f"{붕어빵}의 원하시는 개수를 입력해주세요. "))
        재고[붕어빵] += 개수
        new[붕어빵] += 개수
```

__[문제 6: 매출 확인]__

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
order = {"팥붕어빵": 0, "슈크림붕어빵": 0, "초코붕어빵": 0}
menu = {"1": "팥붕어빵", "2": "슈크림붕어빵", "3": "초코붕어빵"}
가격 = {"팥붕어빵": 1000, "슈크림붕어빵": 1200, "초코붕어빵": 1500}
total = 0
주문내역 =[]

while True:
    주문 = input("주문하실 상품의 번호를 선택해주세요. (1. 팥붕어빵, 2. 슈크림붕어빵, 3. 초코붕어빵, 4. 종료): ")
    if 주문 in menu:
        붕어빵 = menu[주문]
        개수 = int(input(f"{붕어빵}의 원하시는 개수를 입력해주세요: "))
        order[붕어빵] += 개수
        주문가격 = 개수 * 가격[붕어빵]
        total += 주문가격
        주문내역.append(f"{붕어빵} {개수}개({주문가격}원)")
    
    if 주문 == "4":
        print(f"주문하신 {주문내역} 나왔습니다.")
        print("감사합니다.")
        break
print(f"총 매출은 {total}원 입니다.")
```
